```python
import numpy as np
import matplotlib.pyplot as plt
from mfblp import mfblp_filter

%matplotlib inline
```

```python
chr1 = np.load('chr1.npy')
chr1_nonan = chr1[np.isnan(chr1) == False]
chr1_nonan = chr1_nonan[chr1_nonan < 10]
```

```python
s = np.arange(13) / 2.0
f = mfblp_filter(w=7, s=s, a=.15, b=10.0)
m = f.apply_restrict(chr1_nonan)
```

```python
x1 = 2900
x2 = 3500

plt.figure(figsize=(14,8), dpi=300)
plt.plot(chr1_nonan[x1:x2], color='b', alpha=0.5, linewidth=1, label='Raw signal')
plt.plot(m[x1:x2], color='k', alpha=1.0, linewidth=2, label='MFLP filtered')
plt.legend()
plt.show()
```

```python
chr12 = np.load('chr12.npy')
chr12_nonan = chr12[np.isnan(chr12) == False]
chr12_nonan = chr12_nonan[chr12_nonan < 10]
```

```python
s = np.arange(13) / 2.0
f = mfblp_filter(w=3, s=s, a=5., b=10.0)
m = f.apply(chr12_nonan)
```
